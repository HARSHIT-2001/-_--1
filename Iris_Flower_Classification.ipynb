### ~ At first, Importing all the necessary libraries...

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

### ~ Uploading the provided Datasets...

```python
data=pd.read_csv("/content/Iris.csv")
data.head()
```

```python
X = data.drop('Species', axis=1).values
y = data['Species'].values
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()

X1 = dataset.data
y1 = dataset.target
```

```python
X1.shape
y1.shape
```

```python
plt.plot(X1[:, 0][y1 == 0] * X1[:, 1][y1 == 0], X1[:, 2][y1 == 0] * X1[:, 3][y1 == 0], 'r.', label="Satosa")
plt.plot(X1[:, 0][y1 == 1] * X1[:, 1][y1 == 1], X1[:, 2][y1 == 1] * X1[:, 3][y1 == 1], 'g.', label="Virginica")
plt.plot(X1[:, 0][y1 == 2] * X1[:, 1][y1 == 2], X1[:, 2][y1 == 2] * X1[:, 3][y1 == 2], 'b.', label="Versicolour")
plt.legend()
plt.show()
```

### ~ Preprocessing and training of the Datasets...

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.02, random_state=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
model = DecisionTreeClassifier()
print(model.fit(X_train, y_train))
```

```python
X_test = scaler.transform(X_test)
y_pred = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```
